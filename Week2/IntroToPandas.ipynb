- Data is divided into numeric and categorical data.
- Numeric is divided into continuous and discrete.
- Categorical is divided into ordinal, nominal and date/time.
- Ordinal data is in a particular order such as days of the week.
- Nominal data is any unique text-form data such as types of jobs.

Dealing with missing data can be done in a number of ways:
1. Replace with mean
2. Replace with median
3. Create another column of boolean values to indicate missing data values
4. Last resort is to delete the row

Cardinality is the number of unique values in a field such as gender where the cardinality is 2

Label encoding is giving a unique value a number between 1 and the cardinality of the field on a new column

One hot encoding - based on the cardinality of a field, new columns are created to represent the unique values. Here, 1 is used to represent a unique value in that row and 0 for other values
