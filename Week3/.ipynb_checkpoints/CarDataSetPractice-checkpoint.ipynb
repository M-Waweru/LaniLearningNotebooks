```python
import pandas as pd
```

```python
df = pd.read_csv('car_dataset.csv')
```

```python
df.tail()
```

```python
import numpy as np

df.replace('?', np.nan, inplace=True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

Prints the number of missing and available values for each column using a for loop

missing_data.columns.values.tolist() - gets the column names and puts them in a list to be iterated through

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

We shall now get the averages for the following columns and replace the NaN values with the average

```python
# This gets the average of the values in the column

avg_1 = df['normalized-losses'].astype('float').mean(axis = 0)
print(avg_1)
```

```python
# This permanently replaces the NaN values with the average value

df['normalized-losses'].replace(np.nan, avg_1, inplace=True)
```

```python
avg_2 = df['bore'].astype('float').mean(axis=0)
print(avg_2)
```

```python
df['bore'].replace(np.nan, avg_2, inplace = True)
```

```python
avg_3 = df['stroke'].astype('float').mean(axis=0)
print(avg_3)
```

```python
df['stroke'].replace(np.nan, avg_3, inplace=True)
```

```python
avg_4 = df['horsepower'].astype('float').mean(axis=0)
print(avg_4)
```

```python
df['horsepower'].replace(np.nan, avg_4, inplace=True)
```

```python
avg_5 = df['peak-rpm'].astype('float').mean(axis=0)
```

```python
df['peak-rpm'].replace(np.nan, avg_5, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
# idxmax calculates the most common type

df['num-of-doors'].value_counts().idxmax()
```

```python
# Replaces all the np.nan values in the num-of-doors column with Missing

df['num-of-doors'].replace(np.nan, 'Missing', inplace=True)
```

```python
# Drops all rows with NaN in the price column
df.dropna(subset=['price'], axis=0, inplace = True)

# Resets the row indexes since we dropped 4 rows
df.reset_index(drop = True, inplace = True)
```

```python
df.head()
```

```python
df.price.isnull().sum()
```

# Data Normalization

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
df.sample(5)
```

```python
df['length'].mean()
```

```python
df['height'] = df['height']/df['height'].max()
```

# Binning

```python
df['horsepower'] = df['horsepower'].astype(float, copy=True)
```

```python
binwidth = (max(df['horsepower'])-min(df['horsepower']))/4
```

```python
bins = np.arange(min(df['horsepower']), max(df['horsepower']), binwidth)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)
df[['horsepower', 'horsepower-binned']].head(20)
```
